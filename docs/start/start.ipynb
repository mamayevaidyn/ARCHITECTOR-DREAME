The previous code failed with a `KeyError: 'Adj Close'` because, with the latest `yfinance` version and `auto_adjust` defaulting to `True`, the adjusted closing prices are provided under the 'Close' column, not a separate 'Adj Close' column, especially when downloading data for multiple tickers. The correct way to access these prices from the MultiIndex DataFrame is to select the 'Close' level.



```
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

# 1. Define a list of stock tickers
tickers = ['AAPL', 'MSFT', 'BABA', 'ASML']

# 2. Set the period for data collection (last 2-3 years)
end_date = datetime.now()
start_date = end_date - timedelta(days=3 * 365) # Approximately 3 years

print(f"Downloading data for tickers: {tickers}")
print(f"From {start_date.strftime('%Y-%m-%d')} to {end_date.strftime('%Y-%m-%d')}")

# 3. Use yfinance to download historical 'Close' prices
# With auto_adjust=True (default in newer versions), 'Close' column contains adjusted prices.
df = yf.download(tickers, start=start_date, end=end_date)['Close']

# Ensure the DataFrame has tickers as columns and dates as index
# yfinance usually returns data in this format by default for multiple tickers

print("Historical stock data downloaded successfully.")
print("First 5 rows of the DataFrame:")
print(df.head())
```

**Reasoning**:
The previous step successfully downloaded the historical stock prices. The next part of the subtask is to calculate financial metrics, specifically daily returns and the covariance matrix, which are essential for portfolio optimization.



```
import numpy as np

# 1. Calculate daily returns
# The .pct_change() method calculates the percentage change between the current and a prior element.
# It's applied to the DataFrame 'df' to get daily returns for each stock.
daily_returns = df.pct_change().dropna()

# 2. Calculate the annualized covariance matrix of returns
# The .cov() method calculates the covariance matrix.
# We then annualize it by multiplying by 252, which is the approximate number of trading days in a year.
cov_matrix = daily_returns.cov() * 252

# Display the first few rows of daily returns and the covariance matrix
print("Daily Returns (first 5 rows):")
print(daily_returns.head())
print("\nAnnualized Covariance Matrix:")
print(cov_matrix)
```

## Реализация классической оптимизации портфеля

### Subtask:
Разработка кода для трех классических методов оптимизации портфеля: портфель с равными весами, оптимизация среднего и дисперсии Марковица для максимального коэффициента Шарпа, и классическая оптимизация для максимального коэффициента Шарпа с использованием квадратичного программирования.


**Reasoning**:
The first instruction is to calculate the annual expected returns for each asset. This is a prerequisite for the subsequent portfolio optimization steps.



```
import numpy as np

# 1. Calculate annual expected returns
# We'll use the mean of daily_returns and annualize it by multiplying by 252.
annual_expected_returns = daily_returns.mean() * 252

print("Annual Expected Returns:")
print(annual_expected_returns)
```

**Reasoning**:
The next step is to define the helper functions for portfolio calculations (annual return, annual volatility, and Sharpe ratio) as per instructions 2, 3, and 4. After that, I will implement the equal-weighted portfolio strategy by creating the weights, and then calculate and display its annual return, volatility, and Sharpe ratio (instruction 5).



```
import numpy as np
import pandas as pd
from scipy.optimize import minimize

# 2. Define a function to calculate portfolio annual return
def portfolio_annual_return(weights, annual_returns):
    return np.sum(annual_returns * weights)

# 3. Define a function to calculate portfolio annual volatility
def portfolio_annual_volatility(weights, cov_matrix):
    portfolio_variance = np.dot(weights.T, np.dot(cov_matrix, weights))
    return np.sqrt(portfolio_variance)

# 4. Define a function to calculate portfolio Sharpe Ratio
def portfolio_sharpe_ratio(weights, annual_returns, cov_matrix, risk_free_rate=0.01):
    returns = portfolio_annual_return(weights, annual_returns)
    volatility = portfolio_annual_volatility(weights, cov_matrix)
    if volatility == 0:
        return -np.inf # Avoid division by zero, return a very small number
    return (returns - risk_free_rate) / volatility

# 5. For Equal-weighted portfolio:
# a. Create an array of weights where each asset has an equal share.
num_assets = len(annual_expected_returns)
equal_weights = np.array([1./num_assets] * num_assets)

# b. Calculate and print the annual return, annual volatility, and Sharpe ratio for this portfolio.
equal_portfolio_return = portfolio_annual_return(equal_weights, annual_expected_returns)
equal_portfolio_volatility = portfolio_annual_volatility(equal_weights, cov_matrix)
equal_portfolio_sharpe_ratio = portfolio_sharpe_ratio(equal_weights, annual_expected_returns, cov_matrix)

print("\n--- Equal-Weighted Portfolio ---")
print(f"Weights: {equal_weights}")
print(f"Annual Return: {equal_portfolio_return:.4f}")
print(f"Annual Volatility: {equal_portfolio_volatility:.4f}")
print(f"Sharpe Ratio: {equal_portfolio_sharpe_ratio:.4f}")
```

**Reasoning**:
The next step is to implement the classical optimization for the maximum Sharpe Ratio (Markowitz Mean-Variance Optimization) as per instruction 6. This involves defining an objective function to minimize the negative Sharpe ratio, setting constraints for the weights (sum to 1 and individual weights between 0 and 1), and then using `scipy.optimize.minimize` to find the optimal weights and subsequently calculate and print the portfolio's annual return, volatility, and Sharpe ratio.



```
import numpy as np
from scipy.optimize import minimize

# 6. For classical optimization for maximum Sharpe Ratio (Markowitz Mean-Variance Optimization):
# a. Define the objective function to minimize the negative Sharpe Ratio.
def negative_sharpe_ratio(weights, annual_returns, cov_matrix, risk_free_rate=0.01):
    return -portfolio_sharpe_ratio(weights, annual_returns, cov_matrix, risk_free_rate)

# b. Set the constraints for the weights.
# Constraint 1: Sum of weights must be 1.
constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})

# c. Set the bounds for each weight (no short-selling).
num_assets = len(annual_expected_returns)
bounds = tuple((0., 1.) for asset in range(num_assets))

# d. Use scipy.optimize.minimize to find the optimal weights.
# Initial guess for weights can be equal weights.
initial_weights = np.array([1./num_assets] * num_assets)

optimal_results = minimize(
    negative_sharpe_ratio,
    initial_weights,
    args=(annual_expected_returns, cov_matrix),
    method='SLSQP',
    bounds=bounds,
    constraints=constraints
)

optimal_weights_sharpe = optimal_results.x

# e. Calculate and print the optimal weights, annual return, annual volatility, and max Sharpe ratio.
max_sharpe_return = portfolio_annual_return(optimal_weights_sharpe, annual_expected_returns)
max_sharpe_volatility = portfolio_annual_volatility(optimal_weights_sharpe, cov_matrix)
max_sharpe_ratio_value = portfolio_sharpe_ratio(optimal_weights_sharpe, annual_expected_returns, cov_matrix)

print(
"\n--- Classical Optimization for Maximum Sharpe Ratio ---"
)
print(f"Optimal Weights: {optimal_weights_sharpe.round(4)}")
print(f"Annual Return: {max_sharpe_return:.4f}")
print(f"Annual Volatility: {max_sharpe_volatility:.4f}")
print(f"Sharpe Ratio: {max_sharpe_ratio_value:.4f}")
```

**Reasoning**:
The subtask requires implementing classical optimization for the maximum Sharpe Ratio using quadratic programming. To achieve this, I will use the `cvxpy` library, which is a standard tool for convex optimization, including quadratic programming. First, I need to install `cvxpy` to make it available in the environment.



```
print("Installing cvxpy...")
!pip install --quiet cvxpy
print("cvxpy installed.")
```

**Reasoning**:
The next step is to implement the classical optimization for the maximum Sharpe Ratio using Quadratic Programming (QP). I will use `cvxpy` to define the optimization problem, including variables for weights (`y`) and a scaling factor (`k`), the objective function to maximize the modified Sharpe ratio expression, and the necessary constraints. After solving the QP, I will recover the optimal weights and then calculate and display the portfolio's annual return, volatility, and Sharpe ratio.



```
import cvxpy as cp
import numpy as np

# 7. Classical optimization for maximum Sharpe Ratio using Quadratic Programming (QP):
#    This method involves transforming the Sharpe ratio maximization into a QP problem.
#    We introduce a helper variable 'k' and scaled weights 'y = k * w'.

num_assets = len(annual_expected_returns)
risk_free_rate = 0.01

# Define optimization variables
y = cp.Variable(num_assets) # Scaled weights
k = cp.Variable(1)          # Inverse of portfolio volatility

# Objective: Maximize k * (portfolio_return - risk_free_rate)
# This transformation allows the Sharpe ratio maximization to be cast as a QP.
# The expression y.T @ annual_expected_returns.values represents k * w.T @ annual_expected_returns (k * portfolio_return)
# So the objective becomes maximizing k * portfolio_return - k * risk_free_rate
objective = cp.Maximize(y.T @ annual_expected_returns.values - k * risk_free_rate)

# Constraints
constraints = [
    cp.quad_form(y, cov_matrix.values) <= 1, # Constraint related to volatility
    cp.sum(y) == k,                           # Relates scaled weights 'y' to 'k'
    y >= 0,                                   # No short-selling
    k >= 0                                    # k must be non-negative
]

# Define and solve the problem
problem = cp.Problem(objective, constraints)
problem.solve()

# Extract optimal weights
# The optimal weights 'w' are y.value / k.value
optimal_weights_qp = y.value / k.value

# Calculate portfolio metrics using the optimal weights
qp_max_sharpe_return = portfolio_annual_return(optimal_weights_qp, annual_expected_returns)
qp_max_sharpe_volatility = portfolio_annual_volatility(optimal_weights_qp, cov_matrix)
qp_max_sharpe_ratio_value = portfolio_sharpe_ratio(optimal_weights_qp, annual_expected_returns, cov_matrix)

print(
    "\n--- Classical Optimization for Maximum Sharpe Ratio (Quadratic Programming) ---"
)
print(f"Optimal Weights: {optimal_weights_qp.round(4)}")
print(f"Annual Return: {qp_max_sharpe_return:.4f}")
print(f"Annual Volatility: {qp_max_sharpe_volatility:.4f}")
print(f"Sharpe Ratio: {qp_max_sharpe_ratio_value:.4f}")
```

## Бэктестинг классических портфелей

### Subtask:
Оценка производительности классических портфелей на исторических данных (2020-2024 гг.) и расчет ключевых метрик, таких как общая доходность, волатильность и коэффициент Шарпа.


**Reasoning**:
The subtask requires calculating performance metrics for the three classic portfolios. I will start by creating a DataFrame to store the daily returns of each portfolio. Then, for each portfolio, I will calculate its daily returns using the previously derived asset daily returns and the optimal weights. This is a foundational step for all subsequent metric calculations.



```
import pandas as pd
import numpy as np

# Create a DataFrame to store daily returns of each portfolio
portfolio_returns_df = pd.DataFrame(index=daily_returns.index)

# Calculate daily returns for the Equal-Weighted Portfolio
portfolio_returns_df['Equal_Weighted'] = daily_returns.dot(equal_weights)

# Calculate daily returns for the Markowitz Max Sharpe Portfolio
portfolio_returns_df['Markowitz_Max_Sharpe'] = daily_returns.dot(optimal_weights_sharpe)

# Calculate daily returns for the QP Max Sharpe Portfolio
portfolio_returns_df['QP_Max_Sharpe'] = daily_returns.dot(optimal_weights_qp)

print("Daily returns for each portfolio created successfully.")
print("First 5 rows of portfolio_returns_df:")
print(portfolio_returns_df.head())
```

**Reasoning**:
The previous step successfully calculated the daily returns for each portfolio. Now, I will calculate the cumulative return, total return, annualized volatility, annualized return, and Sharpe ratio for each portfolio as per the instructions, and then display these metrics.



```
import numpy as np
import pandas as pd

# Define the risk-free rate used previously
risk_free_rate = 0.01

# Create a dictionary to store performance metrics for each portfolio
portfolio_performance = {}

# Loop through each portfolio in the daily returns DataFrame
for portfolio_name in portfolio_returns_df.columns:
    daily_rets = portfolio_returns_df[portfolio_name]

    # a. Calculate cumulative returns
    # Add 1 to daily returns to get return factors, then use .cumprod()
    cumulative_returns = (1 + daily_rets).cumprod()

    # c. Calculate total return (last value of cumulative returns - 1)
    total_return = cumulative_returns.iloc[-1] - 1

    # d. Calculate annualized volatility
    # Standard deviation of daily returns multiplied by sqrt(252)
    annual_volatility = daily_rets.std() * np.sqrt(252)

    # e. Calculate annualized return
    # Mean of daily returns multiplied by 252
    annual_return = daily_rets.mean() * 252

    # f. Calculate Sharpe Ratio
    # (Annualized Return - Risk-Free Rate) / Annualized Volatility
    sharpe_ratio = (annual_return - risk_free_rate) / annual_volatility

    # Store the calculated metrics
    portfolio_performance[portfolio_name] = {
        'Total Return': total_return,
        'Annual Return': annual_return,
        'Annual Volatility': annual_volatility,
        'Sharpe Ratio': sharpe_ratio
    }

# Convert the dictionary to a DataFrame for better display
performance_df = pd.DataFrame(portfolio_performance).T

print("--- Portfolio Performance Metrics ---")
print(performance_df.round(4))
```
